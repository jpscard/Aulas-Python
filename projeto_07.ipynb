```
# Classe que representa um participante do evento
class Participant:
    def __init__(self, name, age, address, resident_sp, event_time):
        self.name = name  # Nome do participante
        self.age = age  # Idade do participante
        self.address = address  # Endere√ßo do participante
        self.resident_sp = resident_sp  # Indicador se o participante reside em S√£o Paulo (True/False)
        self.event_time = event_time  # Hor√°rio do evento para o participante

# Classe que gerencia o sistema de registro de eventos
class EventRegistrationSystem:
    def __init__(self, max_participants=50):
        self.max_participants = max_participants  # Limite m√°ximo de participantes
        self.participants = []  # Lista para armazenar os participantes registrados

    # M√©todo para registrar um participante
    def register_participant(self, participant):
        if len(self.participants) >= self.max_participants:
            return "Inscri√ß√µes encerradas: o limite de vagas foi atingido."  # Verifica se o limite de participantes foi atingido

        if not self.check_conditions(participant):
            return "A inscri√ß√£o falhou: uma ou mais condi√ß√µes n√£o foram atendidas."  # Verifica se o participante atende √†s condi√ß√µes de inscri√ß√£o

        self.participants.append(participant)  # Adiciona o participante √† lista de registrados
        return "Inscri√ß√£o validada com sucesso."  # Confirma a inscri√ß√£o

    # M√©todo para verificar as condi√ß√µes de inscri√ß√£o
    def check_conditions(self, participant):
        if participant.age < 18:
            return False  # Participante deve ter 18 anos ou mais
        if not participant.resident_sp:
            return False  # Participante deve residir em S√£o Paulo
        if participant.event_time.lower() != 'noite':
            return False  # O evento deve ser no per√≠odo da noite
        return True  # Todas as condi√ß√µes foram atendidas

# Fun√ß√£o principal para executar o sistema de inscri√ß√£o
def main():
    event_system = EventRegistrationSystem()  # Cria uma inst√¢ncia do sistema de registro de eventos

    while True:
        print("Sistema de Inscri√ß√£o para Evento")
        print("Digite as informa√ß√µes do participante:")

        # Coleta as informa√ß√µes do participante
        name = input("Nome: ")
        age = int(input("Idade: "))
        address = input("Endere√ßo: ")
        resident_sp = input("Reside em S√£o Paulo (sim/n√£o): ").lower() == 'sim'
        event_time = input("Hor√°rio do evento (manh√£/tarde/noite): ")

        # Cria um objeto Participant com as informa√ß√µes coletadas
        participant = Participant(name, age, address, resident_sp, event_time)
        # Registra o participante no sistema e exibe o resultado
        result = event_system.register_participant(participant)
        print(result)

        # Pergunta se o usu√°rio deseja cadastrar outro participante
        cont = input("Deseja cadastrar outro participante? (sim/n√£o): ").lower()
        if cont != 'sim':
            break

# Executa a fun√ß√£o principal se o script for executado diretamente
if __name__ == "__main__":
    main()
```
